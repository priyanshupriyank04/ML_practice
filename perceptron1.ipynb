```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
print("All libraries imported")
```

```python
df = pd.read_csv('placement.csv')
print("Csv file read")
print("Shape of the dataset is:",df.shape)

print("Head of the dataset looks like:")
df.head()
```

```python
#Data visualisation 
sns.scatterplot(x=df['cgpa'],y=df['resume_score'],hue=df['placed'])
```

```python
x = df.iloc[:,0:2]  #feature input
y = df.iloc[:,-1]   #output label
```

```python
#Training linear perceptron model
from sklearn.linear_model import Perceptron
p = Perceptron()    #initialise a simple perceptron ML model
p.fit(x,y)  #train the model instance 
```

```python
print("the values of the weights:",p.coef_)
print("the value of the bias:",p.intercept_)
```

```python
from mlxtend.plotting import plot_decision_regions

plot_decision_regions(x.values,y.values,clf=p)
```
